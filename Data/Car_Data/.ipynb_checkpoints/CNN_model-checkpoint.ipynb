```python
#import data
df = pd.read_csv("middle_camera_and_steering.csv")
#checking info
df.info()
```

```python
trainsize = 6500
```

```python
#split in training and testing data
#x holds the image
#y holds the steering input
train = df.iloc[:trainsize,:]
test = df.iloc[trainsize:,:]

#training data
x_train = train["middle_cam"]
y_train = train["steering"]

#testing data
x_test = test["middle_cam"]
y_test = test["steering"]

#reset rows
x_test = x_test.reset_index()
y_test = y_test.reset_index()
```
