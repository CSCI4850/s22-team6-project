### doing actual work

```python
#### creating new dataframe that can hold all the data
combined_df = pd.DataFrame(index=range(len(paths)*2),columns=range(3))
combined_df.shape
```

```python
#set columns
combined_df.columns = ["index","image_path","steering"]
combined_df.head
```

```python
#### first save all the original images to the correct folder, numbered from 0 to len(ofCenterPictures)
for i in range(len(paths)):
    image = cv2.imread(paths[i])
    name = "learning_images//"+ str(i) + ".jpeg"
    cv2.imwrite(name,image)
    print("saved: " , i)
    
    #also write to dataframe
    combined_df.loc[i] = [i,name,steering[i]]
    
```

```python
#check dataframe
combined_df.head()
```

```python
#now saving the flipped images too while also appending to the steering datafile
for i in range(len(paths)):
    image = cv2.imread(paths[i])
    image = cv2.flip(image,1) #flip horizontal
    name = "learning_images//"+ str(i+len(paths)) + ".jpeg"
    cv2.imwrite(name,image)
    print("saved: " , (i+len(paths)))
    
    #also write to dataframe
    combined_df.loc[(i+len(paths))] = [(i+len(paths)),name,(steering[i]*-1)]
```

```python
#check if image is flipped, seems to work
print(combined_df.iloc[0], " | ", combined_df.iloc[len(paths)])
```

```python
#print original image
image = cv2.imread(combined_df["image_path"].iloc[0])
cv2.imshow(str(combined_df["steering"].iloc[0]),image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#print flipped image
flipped = cv2.imread(combined_df["image_path"].iloc[len(paths)])
cv2.imshow(str(combined_df["steering"].iloc[len(paths)]),flipped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# export data to new complete csv file
combined_df.to_csv("augmented_data.csv")
```
